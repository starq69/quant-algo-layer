```python
# OK
#
def ex_gap(high,low):
    '''
    TEST : PASSED

    in  : high, low
    out : [idx, %gap]
          idx = timestamp
          %gap : se > 0 ==> up 
                 se < 0 ==> down

    '''
    _high = high
    _low  = low
    
    def i_gap(idx,h,l):
        if l > _high:
            return [idx, (((l - _high) / _high) * 100.0)]
        elif h < _low:
            return [idx, (((h - _low) / _low) * 100.0)]
        else:
            return None
        
    return i_gap
```

```python
import numpy as np
import pandas as pd
import time

def func(text):
    return text + ' is processed'


def test_one():
    data =pd.DataFrame(columns = ['text'], index = np.arange(0, 100000))
    data['text'] = 'text'

    start = time.time()
    data['text'] = data['text'].apply(func)
    print(time.time() - start)


def test_two():
    data =pd.DataFrame(columns = ['text'], index = np.arange(0, 100000))
    data['text'] = 'text'

    start = time.time()

    for index, row in data.iterrows():
        data.loc[index, 'text'] = row['text'] + ' is processed'

    print(time.time() - start)
```
